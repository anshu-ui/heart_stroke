```python
dummies1 = pd.get_dummies(data.work_type)
dummies2 = pd.get_dummies(data.smoking_status)
```

```python
df = merged
```

```python
df.drop(columns=["id"],axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(df[["bmi"]])
```

```python
df["bmi"] = scaler.fit_transform(df[["bmi"]])
df.head()
```

```python
s1 = scaler.fit(df[["avg_glucose_level"]])
df["avg_glucose_level"] = s1.fit_transform(df[["avg_glucose_level"]])
```

```python
df.drop(columns=["id"],axis=1,inplace=True)
df.head()
```

plt.plot(df[["avg_glucose_level"]])

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns=["stroke"])
y = df.stroke
print(x)
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
dt = DecisionTreeClassifier()
sv = SVC()
rf = RandomForestClassifier()
```

```python
dt.fit(x_train,y_train)
sv.fit(x_train,y_train)
rf.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("dt score",dt.score(x_test,y_test))
print("sv scoer",sv.score(x_test,y_test))
print("rf_score",rf.score(x_test,y_test))
```

```python
y_pred1 = dt.predict(x_test)
print(accuracy_score(y_pred1,y_test))
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_pred1,y_test)
print(report)
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score
```

```python
c1 = cross_val_score(dt,x_train,y_train,cv=10)
print(c1)
print(c1.mean())
```

```python
c2 = cross_val_score(sv,x_train,y_train,cv=10)
print(c2)
print(c2.mean())
```

```python
c3 = cross_val_score(rf,x_train,y_train,cv=10)
print(c3)
print(c3.mean())
```

```python
plt.plot(c1,color="blue")
plt.xticks()
plt.plot(c2,color="red")
plt.plot(c3,color="y")
plt.show()
```
